```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
points = pd.read_csv('data.csv')
print(points)
```

```python
plt.scatter(points['size'], points.price)
plt.show()
```

```python
def loss_function(m,b,points):
  total_error = 0
  for i in range(len(points)):
    x=points.iloc[i]['size']
    y=points.iloc[i]['price']
    total_error += (y-(m*x + b))**2
  return total_error/float(len(points))
```

```python
i_m=0
i_b=0
print(loss_function(i_m, i_b, points))
```

```python
def grad(m1,b1,points,l):
  m2 = 0
  b2 = 0
  n = len(points)
  for i in range(len(points)):
    x=points.iloc[i]['size']
    y=points.iloc[i]['price']
    m2 += -(2/n) * x * (y - m1*x + b1)
    b2 += -(2/n) * (y - m1*x + b1)
  m = m1 - m2*l
  b = b1 - b2*l
  return m,b
```

```python
m = 0
b = 0
l = 0.00000001
epochs = 300
for i in range(epochs):
  if i%50==0:
    print(f"epochs : {i}")
  m,b = grad(m,b,points,l)
plt.scatter(points['size'], points['price'], color = "black")
plt.plot(list(range(1000, 3500)), [m*x + b for x in range(1000,3500)], color = "red")

plt.show()
```

```python
def predict(x, m, b):
    return m * x + b

test_size = 2250
predicted_price = predict(test_size, m, b)
print(f"A house of {test_size} sq ft is predicted to cost: ${predicted_price:,.2f}")
```
